```python
sample=hiv_df[((hiv_df['county_name']=='KISUMU') | (hiv_df['county_name']=='MIGORI') | (hiv_df['county_name']=='KISII') |(hiv_df['county_name']=='TURKANA') |(hiv_df['county_name']=='MOMBASA') | (hiv_df['county_name']=='NAIROBI')|(hiv_df['county_name']=='BUSIA'))] #get the currencies for a certain date
```

```python
# bar graph to show new children with HIV for first seven counties
plt.bar(x=sample['county_name'], height=sample['new_hiv_infections_children_0_14'], alpha=0.8, color = 'red')
plt.title('New Hiv Infections Children For stated Counties')
plt.ylabel('Children HIV level')
plt.xlabel('County Name')
```

```python
# Scatter plot to show new children with HIV for first seven counties
plt.scatter(x=sample['county_name'], y=sample['new_hiv_infections_children_0_14'])
plt.title('Scatter plot of new child HIV infections')
plt.ylabel('New hiv infections children 0 14')
plt.xlabel('County Name')
plt.show()
```

```python
# bar graph to show Number of house hold with orphans for first seven counties
plt.bar(x=sample['county_name'], height=sample['no_of_households_with_an_orphan'], alpha=0.8, color = 'blue')
plt.title('Graph for NO of households with an orphan')
plt.ylabel('No of Orphans')
plt.xlabel('County Name')
```

```python
# Scatter plot to show Number of house hold with orphans for first seven counties.
plt.scatter(x=sample['county_name'], y=sample['no_of_households_with_an_orphan'])
plt.title('Scatter to show Number of house hold with orphans')
plt.ylabel('Number of house hold with orphans')
plt.xlabel('County Name')
plt.show()
```

```python
# bar graph to show poor households with orphans for first seven counties
plt.bar(x=sample['county_name'], height=sample['poor_households_with_an_orphan'], color = 'pink')
plt.title('Graph for poor households with an orphan')
plt.ylabel('No of Orphans in poor household')
plt.xlabel('County Name')
```

```python
# Scatter plot to show show poor households with orphans.
plt.scatter(x=sample['county_name'], y=sample['poor_households_with_an_orphan'])
plt.title('Scatter to show poor households with orphans')
plt.ylabel('poor households with orphans')
plt.xlabel('County Name')
plt.show()
```

```python
# bar graph to show population for first seven counties
plt.bar(x=sample['county_name'], height=sample['total_population'], alpha=1.0, color = 'purple')
plt.title('Graph for total population')
plt.ylabel('Population')
plt.xlabel('County Name')
```

```python
# bar graph to show poe prevention of mother to child transmission
plt.bar(x=sample['county_name'], height=sample['poe_prevention_of_mother_to_child_transmission'], alpha=1.0, color = 'skyblue')
plt.title('poe prevention of mother to child transmission')
plt.ylabel('Mother prevention in percentage')
plt.xlabel('County Name')
```

## LINE GRAPH

```python
#Create plot for one county first


plt.figure(figsize = (6, 4), dpi =100)

#Plot the graph
#plt.plot(x-axis, y-axis)
#use the shorthand to format the lines
plt.plot(sample['county_name'], sample['new_hiv_infections_children_0_14'], label = 'New child HIV infections ', color = 'blue', marker = '.', linestyle = '-')
plt.plot(sample['county_name'], sample['no_of_households_with_an_orphan'], label = 'House with orphans', color = 'red', marker = '.', linestyle = '-')
plt.plot(sample['county_name'], sample['poor_households_with_an_orphan'], label = 'Poor house with orphans', color = 'pink', marker = '.', linestyle = '-')
#plt.plot(hiv_df['county_name'], hiv_df['TURKANA'], label = 'TURKANA', color = 'purple', marker = '.', linestyle = '-')
#plt.plot(hiv_df['county_name'], hiv_df['MOMBASA'], label = 'MOMBASA', color = 'black', marker = '.', linestyle = '-')
plt.plot(sample['county_name'], sample['total_population'], 'g.-', label = 'Population')#using the shorthand
#plt.plot(sample['county_name'], sample['poe_prevention_of_mother_to_child_transmission'], 'y.-', label = 'prevention of mother to child transmission')#using the shorthand


#Fix the ticks on the axes
plt.xticks(sample['county_name'],rotation=30)

#Add title
plt.title('Line Graph showing Children HIV Statistics for seven counties', fontdict={'fontweight':'bold', 'fontsize': 18})

#Add axes labels
plt.xlabel('County Name')
plt.ylabel('HIV related Statistics')

#Add legend
plt.legend()

#Save the visual
#plt.savefig('table_2014_adult_hiv_prevalence_rate_by_county.csv', dpi = 100)

#Show plot
plt.show()
```

```python
#Create plot for one county first


plt.figure(figsize = (6, 4), dpi =100)

#Plot the graph
#plt.plot(x-axis, y-axis)
#use the shorthand to format the lines

plt.plot(sample['county_name'], sample['poe_prevention_of_mother_to_child_transmission'], 'y.-', label = 'prevention of mother to child transmission')#using the shorthand


#Fix the ticks on the axes
plt.xticks(sample['county_name'],rotation=60)
plt.yticks(sample['poe_prevention_of_mother_to_child_transmission'],rotation=0)


#Add title
plt.title('Line Graph showing poe prevention of mother to child transmission for seven counties', fontdict={'fontweight':'bold', 'fontsize': 18})

#Add axes labels
plt.xlabel('County Name')
plt.ylabel('poe prevention of mother to child transmission')

#Add legend
plt.legend()

#Save the visual
#plt.savefig('table_2014_adult_hiv_prevalence_rate_by_county.csv', dpi = 100)

#Show plot
plt.show()
```
